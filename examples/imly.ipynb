# IMLY - Interoperable Machine Learning
---

With IMLY, our goal is to make all existing standard machine learning frameworks(say sklearn, suprislib, pytorch, tensorflow etc) interoperable. That is, one can devlop and train a model, say, using Linear Regression in sklearn, and score it using a TensorFlow server.

In this tutorial, we walk through an example demonstrating one such scenario.
 
*Note: Currently, IMLY's support is limited to sklearn's Linear regression and Logistic regression algorithms. Algorithms and APIs from other packages will soon be added in the coming days.*

## Usage
---
Setting the context(Terminologies used) -  
1) Primal model - Primal model refers to the base model provided by the user. For example, the primal model in the scenario demonstrated below would be the `LogisticRegression()` class instance from sklearn.  
2) dope - The dope function imported from imly converts your primal model to it's dnn equivalent. Also, dope ensures that the functional and behavioural aspects of your primal model is retained when it's "dope"d.


*Note - The usage of IMLY is pretty straightforward as long as the user has a decent understanding of basic Sklearn and Keras functionalities.*

## Step 1: Loading and preprocessing dataset
---
In this example we will use the iris dataset. The primal model used here is sklearn's Logistic Regression class. The `dope` function from IMLY converts sklearn's Logistic Regression model to it's Neural Network equivalent.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets

iris = datasets.load_iris()

X = iris.data
Y = iris.target

# Split the data in to test and train batches
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.60, random_state=0)
```

## Step 2: Instantiate the primal model
---
Instantiate the model you wish to convert in to a Neural network. Here, we use sklearn's logistic regression.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

## Step 3: "Dope" your primal model!
---
The `dope` function from IMLY lets you convert your primal model.

```python
from mlsquare.imly import dope
m = dope(model)
```

---
__Note - The warning message you see about redis server is a part of the optimization process IMLY does. The details about this will be covered in the upcoming tutorials(yet to be published). So fret not! These warning messages can be safely ignored.__

## Step 4: Voila! You have successfully Doped your model
---
Once you have successfully run the `dope` function by passing your primal model, the returned model(the variable `m` here) would behave like any other sklearn models. The only difference being that the model is not a standard sklearn model but a dnn equivalent of the model provided by you.

The below mentioned methods demonstrate the resemblance of an IMLY model with sklearn models.

```python
## Fit your model ##
m.fit(x_train, y_train)
```

```python
## Score your model ##
m.score(x_test, y_test)
```

```python
## Save your model ##
m.save('demo_path')
```

*Note - The save method expects a single argument - filename. You will be able to find the saved model in the directory you're running your script from. The model by default is saved in three formats - h5, onnx and a serialized pickle file.*
